### Partimos de la base de la Cobb Douglas :
# $$Q=K^αL^β$$
*Forma Funcional General*
### $$ In(y) = \beta_1 +  \beta_2 In(X) + \varepsilon_i$$
*Forma adaptada al contexto del ejercicio*
# $$ln(PIB) = \beta_1 +  \beta_2 ln(Empleo) + \beta_3 ln(FBKF) + \varepsilon_i$$

### **La forma funcional de la Cobb Douglas emplea logaritmos naturales en donde en este caso particular asignamos las variables de la siguiente forma**

$$PIB = Q(K,L) = Función  de  producción$$

$$Empleo = L = Trabajo$$


$$FBKF = K = Capital$$


## Primeramente utilizamos la librería numpy para crear nuevas columnas que contengan a las variables pero con logaritmo natural

```python
datos["LPIB"] = np.log(datos["PIB"])
datos["LEmpleo"] = np.log(datos["Empleo"])
datos["LFBKF"] = np.log(datos["FBKF"])
datos.head()
```

```python
fig, (ax1 , ax2, ax3) = plt.subplots(3,1, figsize = (30,30))

ax1.plot(datos.index, datos['LPIB'], '-o',  color = 'purple', label = 'LPIB')
ax1.set(xlabel= 'Trimestres', ylabel= "LProducción")
ax1.set_title('PIB')
ax1.legend(loc='best')


ax2.plot(datos.index, datos['LEmpleo'], '-o',  color = 'orange', label = 'LEmpleo')
ax2.set(xlabel= 'Trimestres', ylabel= 'LPersonas empleadas')
ax2.set_title('Empleo')
ax2.legend(loc='best')


ax3.plot(datos.index, datos["LFBKF"],'-o',  color =  "firebrick", label = 'LFBKF')
ax3.set(xlabel= 'Trimestres', ylabel= 'L Formación Bruta de Capital')
ax3.set_title('FBKF')
ax3.legend(loc='best')

plt.subplots_adjust(top=0.84)

plt.show()
```

##  Asignamos aquellas variables explicatorias y dependientes

```python

Y = datos["LPIB"]
X = datos[["LEmpleo", "LFBKF"]]
X = sm.add_constant(X) # procedimental para la constante ( B1  )
#
X.head()# Las observamos 
```

## Empleamos el método MCO y estimamos los coeficientes/parámetros

```python
modelo = sm.OLS(Y,X).fit()
print(modelo.summary())
```

## Probar la existencia de rendimientos constantes de escala

### En el contexto de una función de producción Cobb-Douglas, la hipótesis rendimientos constante de escala implica que la suma de los coeficientes de las variables de entrada (alpha y beta) debe ser igual a 1. Esto indica que si duplica las entradas, la salida será proporcional a 1, también el doble, confirmando los rendimientos constantes a escala.
### $$Q=K^αL^β$$
### $$Rendimientos Constantes de Escala = α + β = 1 = β1In(Empleo) + β2In(FBKF)$$
- Donde β1 y β2 son los coeficientes de las variables de entrada Empleo y FBKF, respectivamente.

### Combinación lineal de coeficientes: en un modelo de regresión lineal, los coeficientes representan el cambio en la variable dependiente para un cambio de una unidad en la variable predictora correspondiente. La hipótesis Rendimientos Constantes de Escala establece que la suma de estos coeficientes debe ser igual a 1.
#### -Una de las representaciones matriciales existentes de la restricción anteriormente discutida es :
### $$[β1, β2] X [Empleo, FBKF]^T = 1$$
#### - Ahora si incluimos todos los coeficientes ( β0, β1, β2 ), constante y pendientes
### [1] $$[0, 1, 1] X [β0, β1, β2]^T = 1$$
### [2] $$0β0 + 1β1 + 1β2 = 1$$

#### - Prueba de Wald: La prueba de Wald es una prueba de hipótesis que evalúa si se cumple un conjunto de restricciones lineales sobre los coeficientes del modelo. La prueba compara la diferencia entre los modelos sin restricciones y restringidos y determina si las restricciones son estadísticamente significativas.

#### - Al especificar la matriz de restricción [0, 1, 1] en la prueba de Wald, esencialmente está probando si la suma de los coeficientes de Empleo y FBKF en el modelo de regresión múltiple es igual a 1. Si el valor p de la prueba de Wald es menor que un nivel de significación elegido (p. ej., 0,05), rechazaría la hipótesis nula y concluiría que los coeficientes no suman 1, lo que indica una desviación de los rendimientos constantes a escala.


```python
restricción_matricial = np.array([[0, 1, 1]])
```

### Establecemos las Hipótesis Nula y Alternativa

### $$[H0] : α + β = 1$$
### $$[H1] : α + β ≠  1$$

#### - Realizamos el wald test como tal

```python
wald_test = modelo.wald_test(restricción_matricial)
print("Wald Test Results:")
print(wald_test)
```

###  Ergo existe una desviación de los rendimientos constantes de escala tanto que el p-value es menor que 0.05 , se descarta la H0 y se acepta la H1
